```python
# Column Drop Second Method
df_new2=df.drop(columns=['Overview','Meta_score'])
df_new2.head()
```

```python
#Row Drop
df_new3=df.drop([0])
df_new3
```

```python
df_filtered=df[df['IMDB_Rating']>9]
df_filtered
```

```python
#Create a new Column 
f=lambda x:(x["Runtime"].split())[0]
df["RuntimeNew"]=df.apply(f,axis=1) 
df.head()
```

```python
# RuntimeNew Column is still object so we need to convert all the column to the numeric
df['RuntimeNew']=df['RuntimeNew'].astype('int')
df_filtered=df.query('RuntimeNew>=250')
df_filtered
```
