```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

df=pd.read_csv("polynomial.csv",sep=";")

polynomial_regression=PolynomialFeatures(degree=4)
```

```python
df
```

```python
plt.scatter(df['level'],df['salary'])
plt.xlabel('Job Level')
plt.ylabel('Salary')
plt.savefig('1.png',dpi=300)
plt.show()
```

```python
# If we use linear regression to predict salary
reg=LinearRegression()
reg.fit(df[['level']],df['salary'])

plt.xlabel('Job Level')
plt.ylabel('Salary')

plt.scatter(df['level'],df['salary'])

xAxis=df['level']
yAxis=reg.predict(df[['level']])
plt.plot(xAxis, yAxis,color="green",label="linear regression")
plt.legend()
plt.show()
```

```python
polynomial_regression=PolynomialFeatures(degree=4)

x_polynomial=polynomial_regression.fit_transform(df[['level']])
```

```python
reg=LinearRegression()
reg.fit(x_polynomial,df['salary'])
```

```python
y_head=reg.predict(x_polynomial)
plt.plot(df['level'],y_head,color='red',label="Polynomial Regression")
plt.legend()
plt.scatter(df['level'],df['salary'])
plt.show()
```

```python
# if new position level added our business
x_polynomial_new=polynomial_regression.fit_transform([[4.5]])
reg.predict(x_polynomial_new)
```
