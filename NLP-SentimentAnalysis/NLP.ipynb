```python
def  process(review):
    # review without HTML tags
    review=BeautifulSoup(review).get_text()
    # review without punctation and numbers
    review=re.sub("[^a-zA-Z]",' ',review)
    # converting into lowercase and splitting to eliminate stopwords
    review=review.lower()
    review=review.split()
    # review without stopwords
    swords=set(stopwords.words("english"))
    review=[word for word in review if word not in swords]
    # join
    return(' '.join(review))
```

```python
# Clean all df with the process function
train_x_all=[]
for r in range(len(df["review"])):
    if (r+1)%1000==0:
        print("No of reviews processed =",r+1)
    train_x_all.append(process(df["review"][r]))
```

```python
x=train_x_all
y=np.array(df["sentiment"])

train_x,test_x,train_y,test_y = train_test_split(x, y, test_size=0.1,random_state=42)
```

```python
# Create Bag of Words maximum 5000 words(column)
vectorizer=CountVectorizer(max_features=5000)
train_x=vectorizer.fit_transform(train_x)
```

```python
# Create Random Forest Model
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(train_x,train_y)
```
