```python
import numpy as np 
from PIL import Image 
from IPython.display import Image as show_image 
from keras.applications.inception_resnet_v2 import InceptionResNetV2 
from keras.preprocessing import image 
from keras.applications.inception_resnet_v2 import preprocess_input,decode_predictions 
import warnings
warnings.filterwarnings('ignore')
```

```python
img=Image.open("sportscar.jpg").resize((299,299))
```

```python
img=np.array(img)
```

```python
img.shape
```

```python
print(img.ndim)
```

```python
img=img.reshape(-1,299,299,3) #We have to add extra dimension to our array so will reshape it
```

```python
img.shape
```

```python
print(img.ndim)
```

```python
img=preprocess_input(img)
```

```python
incresv2_model=InceptionResNetV2(weights='imagenet',classes=1000)
```

```python
print(incresv2_model.summary())
print(type(incresv2_model))
```

```python
show_image(filename='sportscar.jpg')
```

```python
preds=incresv2_model.predict(img)
print('Predicted Category:',decode_predictions(preds,top=2)[0])
```
